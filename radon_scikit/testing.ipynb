```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Specify the file path
sinogram_path1 = os.path.join(os.pardir, 'dataset_foam', 'x_train_sinograms.npy')
sinogram_path2 = os.path.join('dataset_foam', 'x_train_sinograms.npy')

# Load the .npy file
sinogram_data1 = np.load(sinogram_path1)
sinogram_data2 = np.load(sinogram_path2)


# Access and use the loaded data
print("original: ",sinogram_data1.shape)
print("edited: ",sinogram_data1.shape)
```

```python
import numpy as np

sinogram_images1 = sinogram_data1[:5]
sinogram_images2 = sinogram_data2[:5]  # Assuming sinogram_data2 is the data for sinogram 2

# Plot the images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

for i, image in enumerate(sinogram_images1):
    axes[0, i].imshow(image)
    axes[0, i].axis('off')

for i, image in enumerate(sinogram_images2):
    rotated_image = np.rot90(image, k=1)  # Rotate the image by 90 degrees clockwise
    axes[1, i].imshow(rotated_image)
    flipped_image = np.flipud(rotated_image)  # Flip the image upside down
    axes[1, i].imshow(flipped_image)
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```
